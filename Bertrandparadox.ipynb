```python
import random
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import csv
import seaborn as sns
gender=[]
race_or_ethnicity=[]
parental_level_of_education=[]
lunch=[]	
test_preparation_course=[]	
math_score=[]	
reading_score=[]
writing_score=[]
math_score_female=[]
math_score_male=[]
reading_score_female=[]
reading_score_male=[]
df = pd.read_csv('StudentsPerformance.csv')

for index, row in df.iterrows():
   math_score.append(row['math score'])	
   reading_score.append(row['reading score'])
   writing_score.append(row['writing score'])
   gender.append(row['gender'])
   race_or_ethnicity.append(row['race/ethnicity'])
   parental_level_of_education.append(row['parental level of education'])
   lunch.append(row['lunch'])
   test_preparation_course.append(row['test preparation course'])
    
    
```

```python
 
#1-A
plt.scatter(math_score,writing_score, linewidth=0.10)
plt.title('comparison')
plt.xlabel('math score')
plt.ylabel('writing_score')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.show()


```

```python
#1-B
plt.scatter(writing_score,reading_score , linewidth=0.10)
plt.title('comparison')
plt.xlabel('writing score')
plt.ylabel('reading score')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.show()

```

```python
#1-C
n=0


for index, row in df.iterrows():
    if n==0 :
        math_score_female=[]
        math_score_male=[]
        reading_score_female=[]
        reading_score_male=[]
        n=n+1
    
    if row['gender'] == 'female' :
        math_score_female.append(row['math score'])
        reading_score_female.append(row['writing score'])
    else:
        math_score_male.append(row['math score'])
        reading_score_male.append(row['writing score'])
n=0
  
print(len(math_score_female))
print(len(math_score_male))
        
          
plt.scatter(math_score_female,reading_score_female , marker='o', linewidth=0.10, color ='purple', label="girls")
plt.scatter(math_score_male,reading_score_male , marker='o', linewidth=0.10, color ='blue', label="boys")
plt.title('comparison')
plt.xlabel('writing score')
plt.ylabel('reading score')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.show()


ax=sns.stripplot( reading_score,gender,  jitter=True)
ax.set(xlabel=" reading_score", ylabel = "gender")
sns.despine()
plt.show()
ay=sns.stripplot( writing_score,gender,  jitter=True)
ay.set(xlabel=" writing_score", ylabel = "gender")
sns.despine()


plt.show()


```

```python


for n in range(0 , 1000):
    if "master's degree"==parental_level_of_education[n]:
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='green', alpha=0.7 )
    if parental_level_of_education[n]=="bachelor's degree":
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='blue', alpha=0.7)
    if "some college"==parental_level_of_education[n]:
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='purple',alpha=0.7)
    if "high school"==parental_level_of_education[n]:
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='red',alpha=0.7)
    if parental_level_of_education[n]=="high school":
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='pink',alpha=0.7)
    if "some highcollege"==parental_level_of_education[n]:
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='grey', alpha=0.7)

plt.title('comparison')
plt.xlabel('writing score')
plt.ylabel('reading score')
plt.show()


ay=sns.stripplot(writing_score , parental_level_of_education, jitter=True)
ay.set(xlabel=" writing_score", ylabel = "parental_level_of_education")
plt.show()




ax=sns.stripplot(reading_score , parental_level_of_education, jitter=True)
ax.set(xlabel=" reading_score", ylabel = "parental_level_of_education")
plt.show()
```

```python
#2-a
for n in range(0 , 1000):
    if "completed"==test_preparation_course[n]:
        plt.scatter(math_score[n],writing_score[n] , marker='o', linewidth=0.10, color ='red', alpha=0.7 )
    if test_preparation_course[n]=="none":
        plt.scatter(math_score[n],writing_score[n] , marker='o', linewidth=0.10, color ='blue', alpha=0.7)

plt.title('comparison\n blue for none \n red for completed')
plt.xlabel('math_score')
plt.ylabel('writing_score')
plt.show()


```

```python
#2-b
for n in range(0 , 1000):
    if "standard"==lunch[n]:
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='red', alpha=0.8 )
    if lunch[n]=="free/reduced":
        plt.scatter(writing_score[n],reading_score[n] , marker='o', linewidth=0.10, color ='blue', alpha=0.8)

plt.title('comparison\n blue for free/reduced lunch \n red for  standard lunch')
plt.xlabel('writing_score')
plt.ylabel('reading_score')
plt.show()
```

```python
#5-1

sum=0
for n in range (len(writing_score)):
    sum = writing_score[n] + sum
mean=sum/len(writing_score)
sum=0
for n in range (len(writing_score)):
    sum+=(writing_score[n]-mean)**2
var=sum/len(writing_score)
print("E(writing_score) is  " + str(mean))
print("var(writing_score) is  "+ str( var))
print("S(writing_score) is  "+ str( var**0.5))
```
